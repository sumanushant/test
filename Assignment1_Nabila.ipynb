## Feature Engineering

### 1. filling null values in a dataset with 0

```python
#Feature Engineering Imputation
#Fyller alle manglende verdier med 0
import pandas as pd
import numpy as np
ruter=pd.read_csv('Ruter_data.csv',encoding = "ISO-8859-1")
ruter = ruter.fillna(0)
```

#### We can see there are no null data in the ruter dataset

### 1. Converting object type to date type format

```python
ruter=pd.read_csv('Ruter_data.csv',encoding = "ISO-8859-1",delimiter= ';')

df = pd.DataFrame(ruter) 
df['Dato']= pd.to_datetime(df['Dato'])
 
#Extracting Year
df['year'] = df['Dato'].dt.year
#Extracting Month
df['month'] = df['Dato'].dt.month
#Extracting the weekday name of the date
df['day_name'] = df['Dato'].dt.day_name()
df['day_name']
```

```python
ruter.info()
```

### 2. Converting categorical data to category data type for One hot encoding

```python
cols = list(ruter.columns)
for i in cols:
    #print("Actual Number = %d" %15)
    print("Unique count for %s is = %d" %(i,ruter[i].nunique()))
```

```python
for col in ['Fylke', 'Område','Kommune','Linjetype','Linjefylke','Linjeretning']:
    ruter[col] = ruter[col].astype('category')
```

```python
ruter.info()
```

### 3. One hot encoding for categorical data

```python
ruter = pd.get_dummies(data = ruter,columns=['Fylke', 'Område','Linjetype','Linjefylke','Linjeretning'],drop_first=True)
```

```python
ruter.head()
```

```python
### Label encoding for Holdeplass_Fra
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
ruter['Holdeplass_Fra']= label_encoder.fit_transform(ruter['Holdeplass_Fra'])
```

### 4. Detecting outliers and removing them

```python
### box plot
import matplotlib.pyplot as plt
import pandas as pd
ruter = pd.read_csv('Ruter_data.csv',delimiter= ';',encoding = "ISO-8859-1")
columns = ['Kjøretøy_Kapasitet', 'Passasjerer_Ombord']
ruter = ruter.reindex(columns=columns)
plt.figure(figsize=(10,8))
ruter.boxplot(['Kjøretøy_Kapasitet', 'Passasjerer_Ombord'])
```

```python
ruter.sort_values(['Kjøretøy_Kapasitet','Passasjerer_Ombord'], ascending=[True, True])
ruter = ruter.reset_index()
```

```python
df1 = ruter[['Kjøretøy_Kapasitet','Passasjerer_Ombord']]
```

```python
df1.sort_values(by = ['Kjøretøy_Kapasitet','Passasjerer_Ombord'],ascending = True)
df1 = df1.reset_index()
```

```python
columns = ['Kjøretøy_Kapasitet', 'Passasjerer_Ombord']
ruter = ruter.reindex(columns=columns)
df1 = ruter[['Kjøretøy_Kapasitet','Passasjerer_Ombord']]
df1.sort_values(by = ['Kjøretøy_Kapasitet','Passasjerer_Ombord'],ascending = True)
df1 = df1.reset_index()
df1 = ruter[['Kjøretøy_Kapasitet','Passasjerer_Ombord']]
Q1 = df1.quantile(0.30)
Q3 = df1.quantile(0.70)
IQR = Q3 - Q1
df = df1[~((df1 < (Q1 - 1.5 * IQR)) |(df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

#### The interquartile range (IQR) is a measure of variability, based on dividing a data set into quartiles. Quartiles divide a rank-ordered data set into four equal parts. The values that divide each part are called the first, second, and third quartiles; and they are denoted by Q1, Q2, and Q3, respectively.



```python
merge = pd.merge(ruter,df, left_index=True, right_index=True)
```
